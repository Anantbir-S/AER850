```python
from ultralytics import YOLO
import matplotlib.pyplot as plt
from PIL import Image
import cv2
import numpy as np
import random

# Step 1: Load the trained YOLOv8 model
model = YOLO('/content/drive/MyDrive/Project 3 Data/runs/detect/Finl_PCB_Detector/AnantbirFinal3/weights/best.pt')

# Step 2: Path to the specific image
image_path = '/content/drive/MyDrive/Project 3 Data/evaluation/ardmega.jpg'

# Step 3: Customize bounding boxes and text display with different colors
def custom_plot(image, results, box_thickness=2, font_scale=0.6, font_thickness=1):
    """
    Draws bounding boxes and labels with custom thickness, font size, and different colors.
    """
    img = np.array(image)  # Convert PIL image to NumPy array (RGB format)
    img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)  # Convert to BGR for OpenCV processing

    # Assign random colors for each class
    class_colors = {cls: [random.randint(0, 255) for _ in range(3)] for cls in range(len(results.names))}

    for box in results.boxes:
        # Extract box coordinates, confidence, and class
        x1, y1, x2, y2 = map(int, box.xyxy[0])  # Box corners
        conf = float(box.conf[0])  # Confidence score
        cls = int(box.cls[0])  # Class index
        label = f"{results.names[cls]} {conf:.2f}"

        # Get color for the current class
        color = class_colors[cls]

        # Draw rectangle
        cv2.rectangle(img, (x1, y1), (x2, y2), color=color, thickness=box_thickness)

        # Add label text
        (text_w, text_h), _ = cv2.getTextSize(label, cv2.FONT_HERSHEY_SIMPLEX, font_scale, font_thickness)
        cv2.rectangle(img, (x1, y1 - text_h - 5), (x1 + text_w, y1), color=color, thickness=cv2.FILLED)
        cv2.putText(img, label, (x1, y1 - 5), cv2.FONT_HERSHEY_SIMPLEX, font_scale, (0, 0, 0), thickness=font_thickness)

    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert back to RGB for Matplotlib
    return img

# Step 4: Process the specific image and display
results = model.predict(source=image_path, save=False, conf=0.25, iou=0.5)
image = Image.open(image_path)

# Custom plot
pred_image = custom_plot(image, results[0], box_thickness=8, font_scale=3, font_thickness=3)

# Display with Matplotlib
plt.figure(figsize=(10, 10))
plt.imshow(pred_image)  # Display the correctly formatted RGB image
plt.axis('off')
plt.title(f"Predictions for {image_path.split('/')[-1]}")
plt.show()
```
